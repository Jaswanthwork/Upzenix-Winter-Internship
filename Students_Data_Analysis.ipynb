```python
import pandas as pd
import numpy as np

students_data = pd.read_csv('/content/students_dataset.csv')
df = pd.DataFrame(students_data)
print(df.head())
```

```python
#Descriptive Analytics

Mean = df['marks'].mean()
print("\n Mean represents the average value of the data")
print(f"Average(Mean) Marks are : ", Mean)

Median = df['marks'].median()
print("\n Median represents the middle value of the data")
print(f"Median Marks are : ", Median)

Mode = df['marks'].mode()
print("\n Mode represents the most repeated value of the data")
print(f"Mode Marks are : ", Mode)

Variance = df['marks'].var()
print("\n Variance represents the spread of the data")
print(f"Variance Marks are : ", Variance)

Standard_Deviation = df['marks'].std()
print("\n Standard Deviation represents the spread of the data")
print(f"Standard Deviation Marks are : ", Standard_Deviation)
```

```python
import matplotlib.pyplot as plt

#Visualizatons of score distributions
plt.hist(df['marks'], bins=10, edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Histogram of Marks')
plt.show()
```

```python
#Visualizations of score distribution among different subjects
plt.bar(df['subject'], df['marks'])
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.title('Barplot of Marks by Subject')
plt.show()

```

```python
#Pie Chart showing marks contribution over subjects
#Firstly we should group the data
df = df.groupby('subject').agg({'marks': 'sum'}).reset_index()
plt.figure(figsize=(8, 8))
plt.pie(df['marks'], labels=df['subject'], autopct='%1.1f%%')
```
